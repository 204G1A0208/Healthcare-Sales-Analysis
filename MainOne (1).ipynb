# 3. Exploratory Data Analysis(EDA)

### 3.1 Demographic Analysis:

```python
# Demographic Analysis
age_distribution = data['Age'].value_counts()
gender_distribution = data['Gender'].value_counts()
blood_type_distribution = data['Blood Type'].value_counts()

# Plotting Age Distribution
plt.figure(figsize=(10, 6))
plt.bar(age_distribution.index, age_distribution.values, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Age Distribution of Patients')
plt.grid(True)
plt.show()
```

```python
# Plotting Gender Distribution
plt.figure(figsize=(6, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution of Patients')
plt.show()
```

```python
# Plotting Blood Type Distribution
plt.figure(figsize=(8, 6))
blood_type_distribution.plot(kind='bar', color='lightgreen', ylim= (1200,1300))
plt.xlabel('Blood Type')
plt.ylabel('Number of Patients')
plt.title('Blood Type Distribution of Patients')
plt.grid(True)
plt.show()
```

```python
# Examine trends in admission types based on age
age_admission_trends = data.groupby(['Age', 'Admission Type'])['Name'].count().unstack()
age_admission_trends.plot(kind='bar', stacked=True, figsize=(12, 8),legend='k')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title('Trends in Admission Types based on Age')
plt.xticks(rotation=90)
plt.legend(title='Admission Type')
plt.grid(True)
plt.show()
```

```python
# Examine trends in admission types based on gender
gender_admission_trends = data.groupby(['Gender', 'Admission Type'])['Name'].count().unstack()
gender_admission_trends.plot(kind='bar', stacked=False, figsize=(8, 6), edgecolor='k')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.title('Trends in Admission Types based on Gender')
plt.xticks(rotation=0)
plt.legend(title='Admission Type')
plt.grid(True)
plt.show()
```

```python
# Analyze medication usage
medication_counts = data['Medication'].value_counts()  # Count occurrences of each medication
top_10_medications = medication_counts.head(10)  # Get top 10 most frequent medications

# Create bar chart (or pie chart if appropriate)
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

# Choose between bar chart or pie chart based on the number of medications:
if len(top_10_medications) <= 20:
  top_10_medications.plot(kind='bar', color='skyblue', edgecolor='black')
  plt.title("Top 10 Most Frequently Used Medications")
  plt.xlabel("Medication")
  plt.ylabel("Count")
  plt.xticks(rotation=45, ha='right')  # Rotate medication names for readability (optional)
else:
  top_10_medications.plot(kind='pie', autopct='%1.1f%%', startangle=140)  # Adjust startangle for better visualization
  plt.title("Distribution of Medication Usage (Top 10)")

plt.tight_layout()
plt.show()

# Analyze medical condition usage
medical_condition_counts = data['Medical Condition'].value_counts()  # Count occurrences of each condition

# Choose visualization based on the number of conditions:
if len(medical_condition_counts) <= 20:
  medical_condition_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
  plt.title("Frequency of Medical Conditions")
  plt.xlabel("Medical Condition")
  plt.ylabel("Count")
  plt.xticks(rotation=45, ha='right')  # Rotate condition names for readability (optional)
else:
  # For a large number of conditions, consider alternative visualizations like histograms or heatmaps
  # depending on the specific patterns you want to explore.
  print("A large number of medical conditions are present. Consider alternative visualizations like histograms or heatmaps for further analysis.")

plt.tight_layout()
plt.show()
```
